```python
#Deriving summary statistics(the sum, mean, median and standard deviation of the scores for each student)
```

```python
# Isolating the columns which can be computed
```

```python
scores= school_data[['math score','reading score','writing score']]
```

```python
#viewing the columns which have numbers and can be computed
```

```python
scores.head()
```

```python
# Dermining the sum, mean, median and standard deviation and creating new columns with the informtion
```

```python
school_data['sum']= scores.sum(axis=1)
school_data['mean']= scores.mean(axis=1)
school_data['median']= scores.median(axis=1)
school_data['standard deviation']= scores.std(axis=1)
```

```python
#importing the raw data for further summary statistics
```

```python
dataset= pd.read_csv("C:\\Users\DELL\Desktop\web\StudentsPerformance data set.csv")
```

```python
#Deriving the count ,sum, mean, median and standard deviation by race and Ethnicity groups
```

```python
statistics_bygroup = pd.pivot_table(dataset,values=['math score','reading score','writing score'], index=['race/ethnicity'], aggfunc=['count','sum','mean','median','std'])
```

```python
statistics_bygroup
```

```python
#Deriving the count ,sum, mean, median and standard deviation by Gender
```

```python
statistics_bygender = pd.pivot_table(dataset,values=['math score','reading score','writing score'], index=['gender'], aggfunc=['count','sum','mean','median','std'])
```

```python
statistics_bygender
```

```python
#Deriving the count ,sum, mean, median and standard deviation by Parent's level of education
```

```python
statistics_byparent = pd.pivot_table(dataset,values=['math score','reading score','writing score'], index=['parental level of education'], aggfunc=['count','sum','mean','median','std'])
```

```python
statistics_byparent
```

```python
#Deriving the count ,sum, mean, median and standard deviation by lunch availability
```

```python
statistics_bylunch = pd.pivot_table(dataset,values=['math score','reading score','writing score'], index=['lunch'], aggfunc=['count','sum','mean','median','std'])
```

```python
statistics_bylunch
```

```python
#Deriving the count ,sum, mean, median and standard deviation by test preparation course
```

```python
statistics_bytestprep = pd.pivot_table(dataset,values=['math score','reading score','writing score'], index=['test preparation course'], aggfunc=['count','sum','mean','median','std'])
```

```python
statistics_bytestprep
```

```python
#The summary tables are concatenated along the rows and saved to a new file called Summarydata
```

```python
pd.concat([statistics_bygender,statistics_bygroup, statistics_byparent,statistics_bylunch,statistics_bytestprep],keys=['statistics by gender','statistics race/ethnicity', 'statistics by parental level of education','statistics by lunch offer','statistics by test preparation course'], axis=0).to_csv(r"C:\\Users\DELL\Desktop\web\Summarydata.csv")
```
