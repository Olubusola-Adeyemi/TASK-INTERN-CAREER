```python
# Creating empty dataframe and lists to store the inputs of new students detail and performance
auto_data =pd.DataFrame()
auto_number=[] 
auto_gender=[] 
auto_race=[] 
auto_parent=[] 
auto_lunch=[] 
auto_test_prep=[] 
auto_math=[] 
auto_reading=[] 
auto_writing=[]

# Defining a function that would receive user input and check if input is valid
def newstudent_data(): 
    
#Block of code to confirm if additional student information should be added. Also to ensure that an invalid input is flagged
    while True: 
        auto_qs= input("will you like to provide a new set of student data for processing? yes/no ").lower()    
        if auto_qs == 'no': 
            break
        if auto_qs!= 'yes' and auto_qs!= 'no':
            print('invalid input')
            continue
        if auto_qs=='yes':
            print("Kindly review the sample table above. The first 6 columns contains all possible inputs.") 
            print("Please fill the new student's informatation using the sample as a guide.Please note that it is case sensitive") 
            print("The scores should be numbers within 0 and 100")
            
        # To receive the count of student to be added. Data validation used to ensure only numbers between 1 and 5000. 
        # Letters and negative numbers are not allowed
            while True:
                auto_add_student = input("How many students information would you like to add? Between 1 and 5000: ")

                # Check if the input is a number
                if auto_add_student.isdigit():
                    add_number = int(auto_add_student)

                    # Check if the number is within the desired range
                    if 1 <= add_number <= 5000:
                        print(f"You entered a valid number: {add_number}")
                        break
                    else:
                        print("Error: Number must be between 1 and 5000.")
                else:
                    print("Error: Please enter a valid number.")
            
            #Looping through a range of count of additional students details to receive input for each student        
            for i in range(int(auto_add_student)):
                
         # To receive student number. data validation set to ensure that input is unique;not previously entered, 
                #should not be a letter, negative or above 50000
                while True:
                    auto_number1 = input("Student Number:'between 1001 and 50000 only' ")
                    # Check if the input is a number
                    if auto_number1.isdigit():
                        s_number = int(auto_number1)

                # To check if the number has been entered before and if it is within the desired range
                        if s_number not in school_data['Student Number'] and 1001<=s_number<=50000:

                            break
                        else:
                            print("Error: Number has been earlier entered or is not valid.Kindly revalidate")
                    else:
                        print("Error: Please enter a valid number.")
                 
            #To receive gender. data validation set to ensure that input is either male or female. Flag any other input as invalid
                while True:
                    auto_gender1= input("gender: male/female ")
                    if auto_gender1 != 'male' and auto_gender1 != 'female':
                        print('Enter a valid input- male/female')
                    else:
                        break
                
            #To receive race/ethnicity input. data validation set to ensure that input is group A-E. Flag any other input as invalid
                while True:
                    auto_race1= input("race/ethnicity group:options available-group A/group B/group C/group D/group E ")
                    if auto_race1 != 'group A' and auto_race1 !='group B'and auto_race1 != 'group C' and auto_race1 !='group D' and auto_race1 !='group E':
                        print('Enter a valid input, note that input is case sensitive')
                    else:
                        break
            
            #To receive parental level of education input. data validation set to ensure that input is 
            # bachelor's degree/ some college/ master's degree/ associate's degree/ high school. Flag any other input as invalid
                while True:
                    auto_parent1= input("Parental level of education:options available-bachelor's degree/some college/master's degree/associate's degree/high school ").lower()
                    if auto_parent1 != "bachelor's degree" and auto_parent1 !='some college'and auto_parent1 != "master's degree" and auto_parent1 !="associate's degree" and auto_parent1 !='high school':
                        print('Enter a valid input')      
                    else:
                        break
            
            #To receive lunch type input. data validation set to ensure that input is 'standard' or 'free/reduced'. 
            #Flag any other input as invalid   
                while True:
                    auto_lunch1= input("lunch type:standard or free/reduced ").lower()
                    if auto_lunch1 != 'standard' and auto_lunch1 !='free/reduced':
                        print('Enter a valid input')
                    else:
                        break
            
            #To receive test preparation type input. data validation set to ensure that input is 'none' or 'completed'. 
            #Flag any other input as invalid
                while True:
                    auto_test_prep1= input("test preparation course:options avaialble -none/completed ").lower()
                    if auto_test_prep1 != 'none' and auto_test_prep1 !='completed':
                        print('Enter a valid input')
                    else:
                        break
            
            #To receive math score input. data validation set to ensure that input is a number and between 0 and 100. 
            #Flag any other as invalid
                while True:
                    auto_math1 = input("math score: Between 0 and 100 ")
                    # Check if the input is a number
                    if auto_math1.isdigit():
                        number = int(auto_math1)
                    # Check if the number is within the desired range
                        if 0 <= number <= 100:
                            pass
                            break
                        else:
                            print("Error: Number must be between 0 and 100.")
                    else:
                        print("Error: Please enter a valid number.")
                
            #To receive reading score input. data validation set to ensure that input is a number and between 0 and 100. 
            #Flag any other as invalid
                while True:
                    auto_reading1 = input("Reading score: Between 0 and 100 ")
                    # Check if the input is a number
                    if auto_reading1.isdigit():
                        r_number = int(auto_reading1)
                    # Check if the number is within the desired range
                        if 0 <= r_number <= 100:
                            pass
                            break
                        else:
                            print("Error: Number must be between 0 and 100.")
                    else:
                        print("Error: Please enter a valid number.")
                
            #To receive writing score input. data validation set to ensure that input is a number and between 0 and 100. 
            #Flag any other as invalid 
                while True:
                    auto_writing1 = input("Writing score: Between 0 and 100 ")
                    # Check if the input is a number
                    if auto_writing1.isdigit():
                        w_number = int(auto_writing1)
                    # Check if the number is within the desired range
                        if 0 <= w_number <= 100:
                            pass
                            break
                        else:
                            print("Error: Number must be between 0 and 100.")
                    else:
                        print("Error: Please enter a valid number.")   
                
                #Appending each student detail in each loop to the empty list defined above
                auto_number.append(auto_number1)
                auto_gender.append(auto_gender1)
                auto_race.append(auto_race1)
                auto_parent.append(auto_parent1)
                auto_lunch.append(auto_lunch1)
                auto_test_prep.append(auto_test_prep1)
                auto_math.append(number)
                auto_reading.append(r_number)
                auto_writing.append(w_number)
        
        # returning the populated information to the defined empty list outside the function
        return auto_number, auto_race, auto_parent, auto_lunch, auto_test_prep, auto_math, auto_reading,auto_writing

# calling the function
newstudent_data()

# creating the columns and information for the Dataframe auto_data
auto_data['Student Number'] = auto_number 
auto_data['gender'] = auto_gender 
auto_data['race/ethnicity'] = auto_race 
auto_data['parental level of education'] = auto_parent 
auto_data['lunch'] = auto_lunch 
auto_data['test preparation course'] = auto_test_prep 
auto_data['math score'] = auto_math 
auto_data['reading score'] = auto_reading 
auto_data['writing score'] = auto_writing

# viewing the full information entered in a table
auto_data

# merging the additional student's data to the imported data. The new information is given on top.
school_data=pd.concat([auto_data, school_data], keys=['Student Number','gender','race/ethnicity', 'parental level of education','lunch type','test preparation course','math score','reading score','writing score'],axis=0, ignore_index= True)

#viewing the updated table
school_data
```
