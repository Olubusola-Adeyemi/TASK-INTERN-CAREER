```python
url  = 'https://www.jumia.com.ng/boys-tops-tees/'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html.parser')
#print(soup)
```

```python
#By observing and inspecting the web page, the following was discovered: 
# a. Each page has about 40 to 45 items and there is a total of about 50 pages, 
# b. There are links within the 'a' HTML tag with HTML class,'core' on the web page which forms part of the 
# characters in the link to the individual items page. e.g '/fashion-boy-polo-dark-red-259348656.html'
# c. Concatenating the prefix 'https://www.jumia.com.ng' with the href element described above gives the actual link 
# to each item's individual page
# The links to each webpage is similar. The difference is in the page number 
#for example 'https://www.jumia.com.ng/boys-tops-tees/?page=(1...)#catalog-listing' for page 1 and
#            'https://www.jumia.com.ng/boys-tops-tees/?page=(2...)#catalog-listing' for page 2 and so on
# d. More information can be scraped from the individual links
```

```python
#creating an empty list "links" to store all links to the individual pages
```

```python
links =[]
```

```python
#Defining a function called linksurla which would loop through all 50 pages available and return all individual item links 
```

```python
def linksurla():
    for b in range(1,51):
        nextlink=  (f'https://www.jumia.com.ng/boys-tops-tees/?page={b}#catalog-listing')
        nextpage = requests.get(nextlink)
        nextsoup = BeautifulSoup(nextpage.text,'html.parser')
        for a in nextsoup.find_all('a',attrs={'core'}, href=True): 
            links_with_text= 'https://www.jumia.com.ng'+a['href']
            links.append(links_with_text)
    return links
```

```python
#To get more information on each item,empty lists are created to store the information that would be scrapped for each url
```

```python
item_name=[]
item_price=[]
seller_name=[]
seller_rating=[]
product_detail=[]
box_content=[]
specification=[]
```

```python
# A function called others is defined to loop through the DataFrame rows and column to derive more information 
# by the tags, ids, classes which was inspected on the items page
```

```python
# the number of items to be scraped can be specified by changing the range inputs. Only 200 items would be scraped below.
```

```python
def others():
    for n in range(0,200):
        access=links1.iloc[n][0]
        eachpage= requests.get(access)
        eachsoup = BeautifulSoup(eachpage.text, 'html.parser')
        try:
            item_name1= eachsoup.find('h1', attrs={'-fs20 -pts -pbxs'}).text.strip()
            item_price1= eachsoup.find('span',attrs={'-b -ubpt -tal -fs24 -prxs'}).text.strip()[2:]
            seller_name1 = eachsoup.find("p", attrs= {"-m -pbs"}).text.strip()
            seller_rating1 = eachsoup.find("bdo", attrs= {"-m -prxs"}).text.strip()
            product_detail1 = eachsoup.find("div", attrs= {"markup -mhm -pvl -oxa -sc"}).text.strip()[:200]
            box_content1 = eachsoup.find("div", attrs= {"markup -pam"}).text.strip()
            specification1 = eachsoup.find("li", attrs= {"-pvxs"}).text.strip() 
        except AttributeError:
            pass
        item_name.append(item_name1)
        item_price.append(item_price1)
        seller_name.append(seller_name1)
        seller_rating.append(seller_rating1)
        product_detail.append(product_detail1)
        box_content.append(box_content1)
        specification.append(specification1)
    return item_name, item_price, seller_name, seller_rating, product_detail, box_content, specification 
```
